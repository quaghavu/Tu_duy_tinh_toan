```python
#w9a1
def mergedssx(lst1, lst2):
  i, j = 0, 0
  lst = []
  while i < len(lst1) and j < len(lst2):
    if lst1[i] < lst2[j]:
      lst.append(lst1[i])
      i += 1
    else:
      lst.append(lst2[j])
      j += 1
  lst.extend(lst1[i:]) #extend: them lan luot phan tu con lai cua ds vao cuoi ds moi
  lst.extend(lst2[j:])
  return lst
lst1 = list(map(int, input().split()))
lst2 = list(map(int, input().split()))
print(*mergedssx(lst1, lst2))
```

```python
#w9a2
def timso(matrix, k):
  lst = []
  for i in matrix:
    lst.extend(i)
  lst.sort()
  return(lst[k - 1])

matrix = []
for i in range(3):
  row = list(map(int, input().split()))
  matrix.append(row)
k = int(input("nhap k < 15: "))
print(timso(matrix, k))
```

```python
#w9a3
def hoandoi(n):
  lst = list(str(n))
  swapped = False
  for i in range(len(lst) - 1):
    vitridoi = lst[i]
    mang = lst[i+1:]
    min_ = min(mang)
    if i == 0 and min_ == "0":
      continue
    if min_ < vitridoi:
      for j in range(len(lst) - 1, i, -1):
        if lst[j] == min_:
          lst[i], lst[j] = lst[j], lst[i]
          swapped = True
          break
    if swapped: break
  if not swapped:
    for k in range(1, len(lst)):
      if lst[-1] != lst[-1 - k]:
        lst[-1], lst[-1-k] = lst[-1-k], lst[-1]
        swapped = True
        break
      if swapped: break
  return int("".join(lst))
n = int(input())
print(hoandoi(n))
```

```python
#w9a7
def CanPlaceFlowers(lst, k):
  count = 0
  n = len(lst)
  for i in range(n):
    if lst[i] == 1:
      continue
    left = (i == 0) or (lst[i - 1] == 0)
    right = (i == n - 1) or (lst[i + 1] == 0)
    if left and right:
      lst[i] = 1
      count += 1
  if count >= k:
    print("True")
  else:
    print("False")

lst = list(map(int, input().split()))
k = int(input())
CanPlaceFlowers(lst, k)
```

```python
#w9a8
def is_path_crossing(str):
  x = 0
  y = 0
  visited = set()
  visited.add((x, y))
  for i in str:
    if i == "R":
      x += 1
    if i == "L":
      x -= 1
    if i == "U":
      y += 1
    if i == "D":
      y -= 1
    if (x, y) in visited:
      return "True"
    visited.add((x, y))
  return "False"

str = input()
print(is_path_crossing(str))
```
