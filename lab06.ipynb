```python
#w6a4 *
s = input("nhap vao key:value : ").split()
dct = {}
for i in s:
  if ":" in i:
    key, value = i.split(":")
    if key not in dct:
      dct[key] = [value]
    else:
      dct[key].append(value)
print(dct)
```

```python
#w6a9 *
import ast
dct1 = ast.literal_eval(input("nhap vao dict 1: ")) # ham chuyen chuoi da nhap thanh dict
dct2 = ast.literal_eval(input("nhap vao dict 2: "))
for i in dct1:
  if i in dct2:
    dct1[i] += dct2[i]
for i in dct2:
  if i not in dct1:
    dct1[i] = dct2[i]
sorted_keys = sorted(dct1.keys())
sorted_dict = {k: dct1[k] for k in sorted_keys}
print(sorted_dict)
```

```python
#w6a10
lst = list(map(int, input().split()))
k = int(input())
s = set()
for i in range(len(lst)):
  for j in range(i + 1, len(lst)):
    if lst[i] + lst[j] == k:
      s.add((lst[i], lst[j]))
print(sorted(list(s)))
```

```python
#w6a12
lst = list(map(int, input().split()))
dct = {}
for i in lst:
  if i in dct:
    dct[i] += 1
  else:
    dct[i] = 1

s = max(dct.values())
s1 = []
for i in dct:
  if dct[i] == s:
    s1.append(i)
print(min(s1))
```

```python
#w6a17
n = int(input("nhap cap cua ma tran vuong: "))
matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)
duongcheochinh = []
duongcheophu = []
for i in range(n) :
  duongcheochinh.append(matrix[i][i])
  duongcheophu.append(matrix[i][n-1-i])
print(duongcheochinh)
print(duongcheophu)
```

```python
#w6a18
n, m, k = map(int, input("nhap so hang, cot, so k: ").split())
matrix = []
for i in range(n):
  row = list(map(int, input().split()))
  matrix.append(row)
# tinh tong cac phan tu tren hang k
for i in range(n):
  if i + 1 == k:
    print(matrix[i])
    print(sum(matrix[i]))
# tinh tong cac phan tu tren cot k
tong = 0
for i in range(m):
  if i + 1 == k:
    for j in range(n):
      tong += matrix[j][i]
print(tong)
```
