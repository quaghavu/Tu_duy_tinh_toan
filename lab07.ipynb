```python
#w7a1
lst = list(map(int, input().split()))
target = int(input())
def binary_search(arr, left, right, target):
  while left <= right:
    mid = (left + right) // 2
    if arr[mid] == target:
      return mid
    elif arr[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return -1
print(binary_search(lst, 0, len(lst) - 1, target))
```

```python
#w7a2
lst = list(map(int, input().split()))
x = int(input())
def count_occurrences(arr, x) -> int:
  dem = 0
  for i in arr:
    if i == x:
      dem += 1
  return dem
print(count_occurrences(lst, x))
```

```python
#w7a3
lst = list(map(int, input().split()))
def bubble_sort(arr):
  n = len(arr)
  for i in range(n):
      swapped = False
      for j in range(0, n - i - 1):
          if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]
            swapped = True
      if not swapped:
          break
  return arr
print(*bubble_sort(lst))
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
print(*selection_sort(lst))
```

```python
#w7a5
lst = list(map(int, input().split()))
x = int(input())
def demsocap(lst, x):
  dem = 0
  for i in range(len(lst)):
    for j in range(i + 1, len(lst)):
      if lst[i] + lst[j] == x:
        dem += 1
  return dem
print(demsocap(lst, x))
```

```python
#w7a11
s = input()
t = str(input())
dem = 0
for i in s:
  if "," == i:
    s = s.replace(",", " ")
  elif ":" == i:
    s = s.replace(":", " ")
  elif ";" == i:
    s = s.replace(";", " ")
  elif "?" == i:
    s = s.replace("?", " ")
  elif "!" == i:
    s = s.replace("!", " ")
  elif "." == i:
    s = s.replace("."," ")
h=s.split()
for i in h:
  if i == t:
    dem += 1
print(dem)
```
