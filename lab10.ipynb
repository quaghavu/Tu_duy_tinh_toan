```python
#w10a1
def devide(a, b):
  try:
    result = a / b
    print(f"{result:.2f}")
  except ZeroDivisionError:
    print("Loi chia khong")

a, b = map(int, input().split())
devide(a, b)
```

```python
#w10a2
def chiso(n):
  try:
    lst = []
    print(f"nhap {n} so nguyen: ")
    for i in range(n):
      s = int(input(f"nhap so thu {i}: "))
      lst.append(s)
    print(lst[n])
  except IndexError:
    print("Loi chi so")

n = int(input("nhap n: "))
chiso(n)
```

```python
#w10a3
def tinhtongbp(n):
  tong = 0
  for i in str(n):
    tong += int(i) ** 2
  return tong
def happynumber(n):
  set1 = set()
  while n != 1 and n not in set1:
    set1.add(n)
    n = tinhtongbp(n)
  return n == 1

def main():
  try:
    a = int(input())
    if happynumber(a):
      print("Yes")
    else:
      print("No")
  except ValueError:
    print("loi kieu du lieu")
main()
```

```python
#w10a5
def chuyen(lst):
  for i in lst:
    try:
      i = int(i)
      print(i)
    except:
      print("khong the chuyen")
lst = [1, 2, 3, "a", 0.5, 5]
chuyen(lst)
```

```python
#w10a6
import math
def canbachai(x):
  try:
    print(f"{math.sqrt(x):.2f}")
  except ValueError:
    print("So am")

x = int(input("Nhap x: "))
canbachai(x)
```

```python
#w10a10
def check():
  try:
    lst = list(map(int, input("nhap N va mang N ptu: ").split()))
    n = lst[0]
    mang = lst[1:]
    if len(mang) == n and len(set(mang)) == len(mang):
      print(sum(mang))
    else:
      print("Mang khong hop le")
  except:
    print("Mang khong hop le")
check()
```
