```python
#w8a1
class Cylinder:
  def __init__(self, r, h):
    self.r = r
    self.h = h
  def Stoanphan(self):
    dtich = 2 * 3.14 * r * r + 2 * 3.14 * r * h
    return dtich
  def thetich(self):
    thetich = 3.14 * r * r * h
    return thetich
r, h = map(float, input("nhap vao ban kinh, chieu cao: ").split())
tru = Cylinder(r, h)
print(round(tru.Stoanphan(),2))
print(round(tru.thetich(),2))
```

```python
#w8a2
class Date:
  def __init__(self, day, month, year):
    self.day = day
    self.month = month
    self.year = year
  def namnhuan(self):
    if (self.year % 400 == 0) or (self.year % 4 == 0 and self.year % 100 != 0):
      return True
  def songaytrongthang(self):
    if self.month in [1, 3, 5, 7, 8, 10, 12]:
      return 31
    elif self.month in [4, 6, 9, 11]:
      return 30
    elif self.month == 2:
      if self.namnhuan():
        return 29
      else:
        return 28
  def hople(self):
    if self.year < 1:
      return False
    if self.month < 1 or self.month > 12:
      return False
    if self.day < 1 or self.day > self.songaytrongthang():
      return False
    return True
  def ngayketiep(self):
    if not self.hople():
      return "INVALID"
    else:
      self.day += 1
      if self.day > self.songaytrongthang():
        self.day = 1
        self.month += 1
        if self.month >12:
          self.month = 1
          self.year += 1
      return f"{self.day}/{self.month}/{self.year}"

day, month, year = map(int, input("nhap (ngay/thang/nam): ").split("/"))
date = Date(day, month, year)
print(date.ngayketiep())
```

```python
#w8a3
import math
class diem:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def kiemtra(self):
    if self.x == 0 and self.y == 0:
      print("M la goc toa do")
    else:
      print("M khong la goc toa do")
    if self.x > 0:
      print("M nam ben phai truc tung")
    else:
      print("M khong nam ben phai truc tung")
    if self.y > 0:
      print("M nam tren truc hoanh")
    else:
      print("M khong nam tren truc hoanh")
  def khoangcach(self):
    d = math.sqrt(self.x ** 2 + self.y **2)
    return d
x, y = map(int, input("nhap hoanh do, tung do M: ").split())
M = diem(x, y)
M.kiemtra()
print(round(M.khoangcach(), 2))
```

```python
#w8a6
class Rectangle:
  def __init__(self, w, h):
    self.w = w
    self.h = h
  def area(self):
    return self.w * self.h
  def perimeter(self):
    return 2 * (self.w + self.h)
  def scale(self, k):
    self.w *= k
    self.h *= k
def main():
  w, h, k = map(float, input().split())
  hcn = Rectangle(w, h)
  hcn.scale(k)
  print(f"{hcn.area():.2f} {hcn.perimeter():.2f}")
main()
```

```python
#w8a7
import math
class Fraction:
  def __init__(self, tu, mau):
    if mau < 0:
      tu = - tu
      mau = - mau
    ucln = math.gcd(tu, mau)
    self.tu = tu // ucln
    self.mau = mau // ucln
  def add(self, other):
    tu = self.tu * other.mau + self.mau * other.tu
    mau = self.mau * other.mau
    return Fraction(tu, mau)
  def sub(self, other):
    tu = self.tu * other.mau - self.mau * other.tu
    mau = self.mau * other.mau
    return Fraction(tu, mau)
  def mul(self, other):
    tu = self.tu * other.tu
    mau = self.mau * other.mau
    return Fraction(tu, mau)
  def div(self, other):
    tu = self.tu * other.mau
    mau = self.mau * other.tu
    return Fraction(tu, mau)
  def __str__(self):
    return f"{self.tu}/{self.mau}"

def main():
  a, b, op, c, d = input().split()
  a, b, c, d = int(a), int(b), int(c), int(d)
  f1 = Fraction(a, b)
  f2 = Fraction(c, d)
  if op == "+":
    print(f1.add(f2).__str__())
  if op == "-":
    print(f1.sub(f2).__str__())
  if op == "*":
    print(f1.mul(f2).__str__())
  if op == "/":
    print(f1.div(f2).__str__())
main()
```

```python
#w8a8
class BankAccount:
  def __init__(self, owner, balance):
    self.owner = owner
    self.balance = balance
  def deposit(self, sotien):
    self.balance += sotien
  def withdraw(self, sotien):
    if self.balance > sotien:
      self.balance -= sotien

def main():
  owner, balance = input().split()
  balance = int(balance)
  account = BankAccount(owner, balance)
  n = int(input())
  for i in range(n):
    s1, s2 = input().split()
    s2 = int(s2)
    if s1 == "WITHDRAW":
      account.withdraw(s2)
    if s1 == "DEPOSIT":
      account.deposit(s2)
  print(f"{account.balance:.2f}")
main()
```

```python
#w8a10
class Book:
  def __init__(self, title, author, year):
    self.title = title
    self.author = author
    self.year = int(year)
class Library:
  def __init__(self):
    self.lst = []
  def add(self,book):
    self.lst.append(book)
  def count_by_author(self, name):
    count = 0
    for i in self.lst:
      if i.author == name:
        count += 1
    return count
  def find_by_year(self, y):
    count = 0
    for i in self.lst:
      if i.year == int(y):
        count += 1
    return count

def main():
  n = int(input())
  library = Library()
  for i in range(n):
    s1, s2 = input().split(" ", 1)
    if ";" in s2:
      title, author, year = s2.split(";")
      year = int(year)
      book = Book(title, author, year)
      library.add(book)
    if s1 == "COUNT":
      print(library.count_by_author(s2))
    if s1 == "COUNTYEAR":
      s2 = int(s2)
      print(library.find_by_year(s2))
main()
```

```python
#w8a11
import math
class Diem:
  def __init__(self, x, y, z):
    self.x = x
    self.y = y
    self.z = z
def vct(p1, p2):
  return Diem(p2.x - p1.x, p2.y - p1.y, p2.z - p1.z)
def vtpt(p1,p2):
  return Diem(p1.y*p2.z-p1.z*p2.y, p1.z*p2.x-p1.x*p2.z, p1.x*p2.y-p1.y*p2.x)
def tichvohuong(p1, p2):
  return p1.x*p2.x+p1.y*p2.y+p1.z*p2.z
def dodaivct(p1):
  return math.sqrt(p1.x**2+p1.y**2+p1.z**2)

lst = []
for i in range(4):
  x, y, z = input().split()
  x, y, z = float(x), float(y), float(z)
  lst.append(Diem(x, y, z))
ab = vct(lst[0], lst[1])
bc = vct(lst[1], lst[2])
cd = vct(lst[2], lst[3])
vtptabc = vtpt(ab, bc)
vtptbcd = vtpt(bc, cd)
arccos = tichvohuong(vtptabc, vtptbcd)/(dodaivct(vtptabc)*dodaivct(vtptbcd))
rad = math.acos(arccos)
degrees = math.degrees(rad)
print(f"{degrees:.2f}")
```

```python
#w8a12
class Complex:
  def __init__(self, re, im):
    self.re = re
    self.im = im
  def tich(self, other):
    ree = self.re * other.re - self.im * other.im
    imm = self.re * other.im + self.im * other.re
    return Complex(ree, imm)
def main():
  re1, im1 = map(float, input().split())
  re2, im2 = map(float, input().split())
  z1 = Complex(re1, im1)
  z2 = Complex(re2, im2)
  print(f"{z1.tich(z2).re} {z1.tich(z2).im}")
main()
```

```python
#w8a13
import math
class Triangle:
  def __init__(self, x, y, z):
    self.x = x
    self.y = y
    self.z = z
  def area(self):
    if (self.x+self.y)>self.z and (self.x+self.z)>self.y and (self.y+self.z)>self.x:
      p = (self.x+self.y+self.z)/2
      s = math.sqrt(p*(p-self.x)*(p-self.y)*(p-self.z))
      return f"{s:.2f}"
    else:
      return "invalid"

x, y, z = map(int, input().split())
tamgiac = Triangle(x, y, z)
print(tamgiac.area())
```
