```python
#w5a1
a, b = map(int, input("nhap vao 2 so a va b: ").split())
def Max_Of_Two(a, b):
  if a < b:
    a, b = b, a
  return a
print(Max_Of_Two(a, b))
```

```python
#w5a2
a, b = map(int, input("nhap vao 2 so a va b: ").split())
def swap(a, b):
  a, b = b, a
  return f"{a} {b}"
print(swap(a, b))
```

```python
#w5a3
import math
a = int(input("a = "))
def songuyento(a):
  if a < 2:
    return "false"
  elif a == 2:
    return "True"
  else:
    for i in range (2, int(math.sqrt(a)) + 1):
      if a % i == 0:
        return "False"
      else:
        return "True"
print(songuyento(a))
```

```python
#w5a4
# Số hoàn hảo (hay còn gọi là số hoàn chỉnh) là một số nguyên dương mà tổng các ước số thực sự của nó (không tính chính nó) bằng chính nó.
n = int(input("n = "))
def sohoanhao(a):
  tong = 0
  if a >= 2:
    for i in range(1, a):
      if a % i == 0:
        tong += i
    if tong == a:
      return True
    return False
  else:
    return False
print(sohoanhao(n))
```

```python
#w5a5
lst = list(map(int, input().split()))
k = int(input("k = "))
def vitrids(lst, k):
  for i in range(len(lst)):
    if lst[i] == k:
      return i + 1
  return -1
print(vitrids(lst, k))
```

```python
#w5a6
n = int(input("n = "))
def giaithua(n):
  s = 1
  for i in range(1, n+1):
    s *= i
  return s
print(giaithua(n))
```

```python
#w5a7
a, operat, b = map(str, input().split())
def calculator(num1, num2, operat):
    result = 0
    if operat == '+':
        result = num1 + num2
    elif operat == '-':
        result = num1 - num2
    elif operat == '*':
        result = num1 * num2
    elif operat == '/':
        if num2 == 0:
            return "Lỗi: Không thể chia cho 0"
        result = num1 / num2
    else:
        return "Lỗi: Toán tử không hợp lệ"
    return round(result, 2)
print(calculator(int(a), int(b), operat))
```

```python
#w5a8
a, b = map(int, input().split())
def khoangcachHamming(a, b):
    s = a ^ b # tim cac vi tri = 1 (cac vi tri khac nhau)
    dem = 0
    while s > 0:
        if s % 2 == 1: # ktra bit cuoi cung co = 1 ko
            dem += 1
        s = s // 2
    return dem
print(khoangcachHamming(a, b))
```

```python
#w5a9
n = int(input("nhap so nguyen duong n= "))
def s(n):
  s = 0
  while n != 0:
    s += (n % 10)
    n //= 10
  return s
print(s(n))
```

```python
#w5a10
a, b = map(str, input("nhap vao 2 chuoi: ").split())
def anhxa(a, b):
    if len(a) != len(b):
        return False
    map_a_b = {}  # Tạo 2 từ điển để lưu ánh xạ 2 chiều: map_a_b = {"i": x}, map_b_a = {"x": i}
    map_b_a = {}
    for i, x in zip(a, b): # ktra tung cap
        if i in map_a_b:
            if map_a_b[i] != x:
                return False
        else:
            if x in map_b_a:
                return False
            map_a_b[i] = x
            map_b_a[x] = i
    return True
print(anhxa(a, b))
```
