```python
#w4a9
n = int(input())
lst = []
def scp(n):
  lst = []
  i = 1
  while True:
    s = i * i
    if s > n:
      break
    else:
      if len(set(str(s))) == len(str(s)): # ko co ki tu trung lap
         lst.append(s)
         i += 1
  return lst
print(scp(n))
```

```python
#w4a10
# Quy tắc tạo ra dãy Collatz từ một số nguyên dương n: Nếu số hiện tại là Chẵn: Chia cho 2 (n / 2). Nếu số hiện tại là Lẻ: Nhân 3 cộng 1 (3n + 1). Lặp lại cho đến khi kết quả là 1.
n = int(input("Nhap n= "))
def tinh_do_dai_collatz(n):
    do_dai = 1
    while n != 1:
        if n % 2 == 0:
            n = n // 2
        else:
            n = 3 * n + 1
        do_dai += 1
    return do_dai
x = 0
y = 0
for i in range(1, n + 1):
  if y < tinh_do_dai_collatz(i):
    y = tinh_do_dai_collatz(i)
    x = i

print(f"{x} {y}")
```

```python
#w4a30
s = input("Nhập chuỗi: ")
hoa = 0
thuong = 0
so = 0
for i in s:
    if i.isupper():
        hoa += 1
    elif i.islower():
        thuong += 1
    elif i.isdigit():
        so += 1
print(f"In hoa: {hoa}")
print(f"In thường: {thuong}")
print(f"Số: {so}")
```
