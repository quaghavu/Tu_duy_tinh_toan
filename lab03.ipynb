```python
#2
# (xor: a ^ a = 0, a ^ 0 = a)
a, b = map(int, input("Nhập a và b: ").split())
if a != b:
  a = a ^ b
  b = a ^ b # a = a ^ b, b = a (cũ)
  a = a ^ b # a = b (cũ)
print(a, b)
```

```python
#16
x = float(input("Nhập vào một số thực: "))
x_nguyen = int(x)
if x > 0 and x != x_nguyen:
    lam_tron_len = x_nguyen + 1
else:
    lam_tron_len = x_nguyen
if x < 0 and x != x_nguyen:
    lam_tron_xuong = x_nguyen - 1
else:
    lam_tron_xuong = x_nguyen
if x >= 0:
    gan_nhat = int(x + 0.5)
else:
    gan_nhat = int(x - 0.5)
print(f"{lam_tron_len} {lam_tron_xuong} {gan_nhat}")
```
